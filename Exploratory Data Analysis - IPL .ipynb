```python
# Decision in every toss either fielding or batting

Toss=mat.toss_decision.value_counts()
labels=np.array(Toss.index)
sizes = Toss.values
colors = ['pink', 'violet']
plt.figure(figsize = (5,3))
plt.pie(sizes, labels=labels, colors=colors,autopct='%1.1f%%', shadow=True,startangle=90)
plt.title('Toss result',fontsize=10)
plt.axis('equal')
plt.show()
```

```python
# What is the probability of winning a match if the toss was won?

win_match = 0
for index, value in mat.iterrows():
    if(value['toss_winner']==value['winner']):
        win_match += 1

print(f'The number of times the team winning toss have won: {win_match}')
prob = win_match/len(mat)
print('The probability of winning if the toss was won: {:.2f}' .format(prob))  
```

```python
#The total runs by fours hit and the total number of fours hit by each team

four_hit=delidata[delidata['batsman_runs']==4]
four_hit.groupby('batting_team')['batsman_runs'].agg([('runs by fours','sum'),('fours','count')])
```

```python
# The fours hit by players

bat_four=delidata.groupby('batsman')['batsman_runs'].agg([('four','count')]).reset_index().sort_values('four',ascending=0)
ax=bat_four.iloc[:10,:].plot('batsman','four',kind='bar',color='pink')
plt.title("Fours hit by players ",fontsize=15)
plt.xticks(rotation=50)
plt.xlabel("Player name",fontsize=10)
plt.ylabel("No of fours",fontsize=10)
plt.show()
```

```python


matdata=mat[['id','season','winner']]

delidata=deli.merge(matdata,how='inner',left_on='match_id',right_on='id')

delidata.columns.values
```

```python
# The number of fours hit in each season

ax=delidata.groupby('season')['batsman_runs'].agg([('four','count')]).reset_index().plot('season','four',kind='bar',color = 'violet')
plt.title("Numbers of fours hit in each season ",fontsize=15)
plt.xticks(rotation=50)
plt.xlabel("Seasons",fontsize=10)
plt.ylabel("No of fours",fontsize=10)
plt.show()
```

```python
# The total runs by the sixes hit and the number of sixes hit by each team

sixes=delidata[delidata['batsman_runs']==6]
sixes.groupby('batting_team')['batsman_runs'].agg([('runs by six','sum'),('sixes','count')])
```

```python
# The number of sixes hit in each season

ax=sixes.groupby('season')['batsman_runs'].agg([('six','count')]).reset_index().plot('season','six',kind='bar',color = 'purple')
plt.title("Numbers of sixes hit in each season ",fontsize=15)
plt.xticks(rotation=50)
plt.xlabel("Seasons",fontsize=10)
plt.ylabel("No of sixes",fontsize=10)
plt.show()
```

```python
# The top 10 leading run scorers in IPL

top10=deli.groupby('batsman')['batsman_runs'].agg(['sum']).reset_index().sort_values('sum',ascending=False).reset_index(drop=True)
top10=top10.rename(columns={'sum':'batsman_runs'})
print("Top 10 Leading Run Scorers in IPL")
top10.iloc[:10,:]
```
